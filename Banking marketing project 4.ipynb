```python
plt.figure(figsize=(20,15),facecolor='white')
plotnumber=1

for col in num_col:
    if plotnumber <=8:
        ax=plt.subplot(2,4,plotnumber)
        sns.distplot(train[col],color = 'm')
        plt.xlabel(col, fontsize=20)
        plt.yticks(rotation = 0, fontsize = 10)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,15),facecolor='white')
plotnumber=1

for col in num_col:
    if plotnumber <=8:
        ax=plt.subplot(2,4,plotnumber)
        sns.boxplot(train[col],color = 'm')
        plt.xlabel(col, fontsize=20)
        plt.yticks(rotation = 0, fontsize = 10)
    plotnumber+=1
plt.show()
```
